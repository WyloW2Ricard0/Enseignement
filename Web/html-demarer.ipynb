# Apprendre le HTML

Le HTML (HyperText Markup Language) est le langage de balisage standard utilisé pour créer des pages web.
Il permet de structurer le contenu d'une page en utilisant des balises qui définissent les différents éléments comme les titres, paragraphes, liens, images, etc.

**Caractéristiques principales :**
- Langage de balisage (pas de programmation)
- Structure le contenu avec des balises (tags)
- Fonctionne avec CSS (style) et JavaScript (interactivité)
- Standard maintenu par le W3C
- Version actuelle : HTML5
