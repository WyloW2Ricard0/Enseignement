# Apprendre le CSS

Le CSS (Cascading Style Sheets) est un langage de feuilles de style utilisé pour décrire la présentation et la mise en forme des documents HTML.

Il permet de contrôler l'apparence des éléments web tels que les couleurs, les polices, les espacements, les dimensions et la disposition des pages.
